```python
columns = ['ID', 'Date', 'Amount', 'Description', 'Service']
df = pd.read_csv('payments_training.tsv', sep='\t', header=None, names=columns)
print(df.head())
```

```python
# Заменяем запятые на точки, удаляем лишние символы
df['Amount'] = df['Amount'].str.replace(',', '.').str.replace('-', '').str.replace('%', '').str.strip()

df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')  
print(df.head())
```

```python
print(df[df['Amount'].isna()])
```

```python
# Удаляем цифры
df['Description'] = df['Description'].str.replace(r'[\d\-]+', '', regex=True)

print(df['Description'])
```

```python
df['Description'] = df['Description'].str.lower()

print(df['Description'].head())
```

```python
df = df.drop(columns=['Date'], errors='ignore')

print(df.head())
```

```python
grouped = df.groupby('Service')['Amount'].sum()
print(grouped)
#сгруппировал данные по service и сумму посчитал
```

```python
# Группируем по категории платежа и считаем средние суммы
category_means = df.groupby('Service')['Amount'].mean().sort_values(ascending=False)

print(category_means)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Столбчатая диаграмма средних сумм платежей по категориям
plt.figure(figsize=(10, 6))
sns.barplot(x=category_means.index, y=category_means.values, palette='viridis')
plt.title('Средняя сумма платежа по категориям')
plt.xlabel('Категория платежа')
plt.ylabel('Средняя сумма платежа')
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy.stats import f_oneway

# Группируем данные по категориям
categories = [df[df['Service'] == category]['Amount'] for category in df['Service'].unique()]

# Проводим дисперсионный анализ ANOVA
anova_result = f_oneway(*categories)

print("p-value ANOVA:", anova_result.pvalue)

if anova_result.pvalue < 0.05:
    print("Есть статистически значимые различия между категориями.")
else:
    print("Статистически значимых различий между категориями нет.")
```

Суммы платежей действительно зависят от категории платежей.
